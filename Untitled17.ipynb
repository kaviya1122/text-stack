```python
import pandas as pd
df = pd.read_csv(r"/content/chandra.csv")
df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'darkgrid', rc = {'figure.figsize':(10, 6)}, font_scale = 1.5)
sns.countplot(data=df, x="name of company")
plt.xticks(rotation=90)
plt.show()
```
